```python
import pandas as pd

# Leer el archivo CSV
df = pd.read_csv('/content/country_vaccinations[1].csv')

# Convertir la columna 'date' al tipo datetime64
df['date'] = pd.to_datetime(df['date'])

# Crear el nuevo DataFrame con los datos originales
conDiferencias = df.copy()

# Calcular la columna derivada "diferencias"
conDiferencias['diferencias'] = conDiferencias['daily_vaccinations'] - conDiferencias['daily_vaccinations_raw']

# Crear el nuevo DataFrame con los datos originales
conCantidad = df.copy()

# Calcular la columna derivada "canVac"
conCantidad['canVac'] = conCantidad['vaccines'].str.split('/').apply(lambda x: len(x))

# Filtrar los registros antes del 20 de diciembre de 2020
antes20 = df[df['date'] < '2020-12-20']

# Filtrar los registros de la vacuna Pfizer
pfizer = df[df['vaccines'].str.contains('Pfizer')]

# Cálculo de la cantidad de vacunas aplicadas por compañía
cantidadPorCompañia = df.groupby('vaccines')['total_vaccinations'].sum().reset_index()

# Cálculo de la cantidad de vacunas aplicadas en todo el mundo
vacunasMundo = df['total_vaccinations'].sum()

# Cálculo del promedio de vacunas aplicadas por país
promedioVacunasPorPais = df.groupby('country')['total_vaccinations'].mean().reset_index()

# Filtrar los registros del día 29/01/21 y calcular la cantidad de vacunas aplicadas en todo el mundo
vacunas29Ene21Mundo = df[df['date'] == '2021-01-29']['total_vaccinations'].sum()

# Crear un objeto ExcelWriter
writer = pd.ExcelWriter('resultadosReto.xlsx')

# Almacenar los dataframes en hojas separadas en el archivo Excel
conDiferencias.to_excel(writer, sheet_name='conDiferencias', index=False)
conCantidad.to_excel(writer, sheet_name='conCantidad', index=False)
antes20.to_excel(writer, sheet_name='antes20', index=False)
pfizer.to_excel(writer, sheet_name='pfizer', index=False)
cantidadPorCompañia.to_excel(writer, sheet_name='cantidadPorCompañia', index=False)
pd.DataFrame({'vacunasMundo': [vacunasMundo]}).to_excel(writer, sheet_name='vacunasMundo', index=False)
promedioVacunasPorPais.to_excel(writer, sheet_name='promedioVacunasPorPais', index=False)
pd.DataFrame({'vacunas29Ene21Mundo': [vacunas29Ene21Mundo]}).to_excel(writer, sheet_name='vacunas29Ene21Mundo', index=False)

# Guardar y cerrar el archivo Excel
writer.save()
writer.close()
```
